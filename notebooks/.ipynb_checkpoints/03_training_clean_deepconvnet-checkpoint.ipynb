```python
# üß† DeepConvNet (version simplifi√©e pour EEG)
class DeepConvNet(nn.Module):
    def __init__(self, n_channels, n_classes, input_time):
        super().__init__()
        self.conv1 = nn.Conv2d(1, 25, (1, 5), padding=(0, 2))
        self.conv2 = nn.Conv2d(25, 25, (n_channels, 1))
        self.batchnorm = nn.BatchNorm2d(25)
        self.activation = nn.ELU()
        self.pool = nn.AvgPool2d((1, 2))
        self.dropout = nn.Dropout(0.5)
        self.classifier = nn.Linear(25 * (input_time // 2), n_classes)
        
    def forward(self, x):
        x = x.unsqueeze(1)  # [N, 1, C, T]
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.batchnorm(x)
        x = self.activation(x)
        x = self.pool(x)
        x = self.dropout(x)
        x = x.view(x.size(0), -1)
        return self.classifier(x)
```

```python
# ‚öôÔ∏è Entra√Ænement cross-validation
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
n_splits = 5
skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

# üßÆ Pond√©ration
class_weights = compute_class_weight('balanced', classes=np.unique(y), y=y)
weights_tensor = torch.tensor(class_weights, dtype=torch.float).to(DEVICE)

for fold, (train_idx, test_idx) in enumerate(skf.split(X_tensor, y_tensor)):
    print(f'
===== Fold {fold+1} =====')
    X_train, X_test = X_tensor[train_idx], X_tensor[test_idx]
    y_train, y_test = y_tensor[train_idx], y_tensor[test_idx]

    train_loader = DataLoader(TensorDataset(X_train, y_train), batch_size=64, shuffle=True)
    model = DeepConvNet(n_channels=64, n_classes=3, input_time=321).to(DEVICE)
    criterion = nn.CrossEntropyLoss(weight=weights_tensor)
    optimizer = optim.Adam(model.parameters(), lr=1e-3)

    # üåÄ Entra√Ænement
    for epoch in range(1, 11):
        model.train()
        running_loss = 0
        correct = 0
        for xb, yb in train_loader:
            xb, yb = xb.to(DEVICE), yb.to(DEVICE)
            optimizer.zero_grad()
            out = model(xb)
            loss = criterion(out, yb)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
            correct += (out.argmax(1) == yb).sum().item()
        acc = correct / len(train_idx)
        print(f'Epoch {epoch:02d} ‚Äî Loss: {running_loss:.4f} ‚Äî Acc: {acc*100:.2f}%')

    # üéØ √âvaluation
    model.eval()
    with torch.no_grad():
        y_pred = model(X_test.to(DEVICE)).argmax(1).cpu()
        print(classification_report(y_test, y_pred, digits=3))
        cm = confusion_matrix(y_test, y_pred)
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
        plt.title(f'Fold {fold+1} ‚Äî Confusion Matrix')
        plt.xlabel('Predicted')
        plt.ylabel('True')
        plt.show()
```

```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import StratifiedKFold
from sklearn.utils.class_weight import compute_class_weight
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

# === Chargement des donn√©es pr√©trait√©es ===
X = np.load("X_windows.npy")
y = np.load("y_windows.npy")

# === Torch tensors ===
X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.long)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
SEED = 42

torch.manual_seed(SEED)
np.random.seed(SEED)

# === Correction des labels pour commencer √† 0 (obligatoire pour CrossEntropyLoss) ===
unique_labels = torch.unique(y)
label_map = {old.item(): new for new, old in enumerate(unique_labels)}
y = torch.tensor([label_map[label.item()] for label in y])
n_classes = len(unique_labels)

# === Poids des classes ===
class_weights = compute_class_weight(class_weight='balanced', classes=np.unique(y), y=y.numpy())
weights_tensor = torch.tensor(class_weights, dtype=torch.float).to(DEVICE)

# === DeepConvNet ===
class DeepConvNet(nn.Module):
    def __init__(self, in_chans, n_classes, input_time_length):
        super().__init__()
        self.conv1 = nn.Sequential(
            nn.Conv2d(1, 25, kernel_size=(1, 5), padding=0),
            nn.Conv2d(25, 25, kernel_size=(in_chans, 1), padding=0),
            nn.BatchNorm2d(25),
            nn.ELU(),
            nn.MaxPool2d(kernel_size=(1, 2)),
            nn.Dropout(0.5)
        )
        self.conv2 = nn.Sequential(
            nn.Conv2d(25, 50, kernel_size=(1, 5)),
            nn.BatchNorm2d(50),
            nn.ELU(),
            nn.MaxPool2d(kernel_size=(1, 2)),
            nn.Dropout(0.5)
        )
        self.conv3 = nn.Sequential(
            nn.Conv2d(50, 100, kernel_size=(1, 5)),
            nn.BatchNorm2d(100),
            nn.ELU(),
            nn.MaxPool2d(kernel_size=(1, 2)),
            nn.Dropout(0.5)
        )
        self.conv4 = nn.Sequential(
            nn.Conv2d(100, 200, kernel_size=(1, 5)),
            nn.BatchNorm2d(200),
            nn.ELU(),
            nn.MaxPool2d(kernel_size=(1, 2)),
            nn.Dropout(0.5)
        )
        with torch.no_grad():
            dummy_input = torch.zeros(1, 1, in_chans, input_time_length)
            out = self._forward_features(dummy_input)
            self.flat_dim = out.view(1, -1).shape[1]

        self.classifier = nn.Linear(self.flat_dim, n_classes)

    def _forward_features(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.conv3(x)
        x = self.conv4(x)
        return x

    def forward(self, x):
        if x.ndim == 3:
            x = x.unsqueeze(1)
        x = self._forward_features(x)
        x = x.view(x.size(0), -1)
        return self.classifier(x)

# === Cross-validation stratifi√©e ===
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)

for fold, (train_idx, test_idx) in enumerate(kf.split(X, y)):
    print(f"\n===== Fold {fold+1} =====")
    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]

    train_loader = DataLoader(TensorDataset(X_train, y_train), batch_size=64, shuffle=True)

    model = DeepConvNet(in_chans=X.shape[1], n_classes=n_classes, input_time_length=X.shape[2]).to(DEVICE)
    criterion = nn.CrossEntropyLoss(weight=weights_tensor)
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

    for epoch in range(1, 11):
        model.train()
        total_loss = 0
        correct = 0
        for xb, yb in train_loader:
            xb, yb = xb.to(DEVICE), yb.to(DEVICE)
            optimizer.zero_grad()
            out = model(xb)
            loss = criterion(out, yb)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
            pred = out.argmax(dim=1)
            correct += (pred == yb).sum().item()
        acc = correct / len(y_train)
        print(f"üß† Fold {fold+1} ‚Äî Epoch {epoch:02d} ‚Äî Loss: {total_loss:.4f} ‚Äî Acc: {acc*100:.2f}%")

    model.eval()
    with torch.no_grad():
        y_pred = model(X_test.to(DEVICE)).argmax(dim=1).cpu()
        print(classification_report(y_test, y_pred, digits=3, zero_division=0))
        cm = confusion_matrix(y_test, y_pred)
        sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
        plt.title(f"Fold {fold+1} ‚Äî Confusion Matrix")
        plt.xlabel("Predicted")
        plt.ylabel("True")
        plt.show()
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import StratifiedKFold
from sklearn.utils.class_weight import compute_class_weight
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

# === Chargement des donn√©es pr√©trait√©es ===
X = np.load("X_windows.npy")
y = np.load("y_windows.npy")

# === Torch tensors ===
X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.long)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
SEED = 42

torch.manual_seed(SEED)
np.random.seed(SEED)

# === Correction des labels pour commencer √† 0 (obligatoire pour CrossEntropyLoss) ===
unique_labels = torch.unique(y)
label_map = {old.item(): new for new, old in enumerate(unique_labels)}
y = torch.tensor([label_map[label.item()] for label in y])
n_classes = len(unique_labels)

# === Poids des classes ===
class_weights = compute_class_weight(class_weight='balanced', classes=np.unique(y), y=y.numpy())
weights_tensor = torch.tensor(class_weights, dtype=torch.float).to(DEVICE)

# === 1. Focal Loss ===
class FocalLoss(nn.Module):
    def __init__(self, gamma=2.0, weight=None):
        super().__init__()
        self.gamma = gamma
        self.weight = weight

    def forward(self, input, target):
        logpt = F.log_softmax(input, dim=1)
        pt = torch.exp(logpt)
        logpt = (1 - pt) ** self.gamma * logpt
        loss = F.nll_loss(logpt, target, weight=self.weight)
        return loss

# === 2. Mixup ===
def mixup_data(x, y, alpha=0.4):
    lam = np.random.beta(alpha, alpha) if alpha > 0 else 1
    batch_size = x.size()[0]
    index = torch.randperm(batch_size).to(x.device)
    mixed_x = lam * x + (1 - lam) * x[index, :]
    y_a, y_b = y, y[index]
    return mixed_x, y_a, y_b, lam

def mixup_criterion(criterion, pred, y_a, y_b, lam):
    return lam * criterion(pred, y_a) + (1 - lam) * criterion(pred, y_b)

# === 3. Temporal Dropout ===
class TemporalDropout(nn.Module):
    def __init__(self, p=0.5):
        super().__init__()
        self.p = p

    def forward(self, x):
        if not self.training or self.p == 0:
            return x
        mask = torch.rand(x.size(0), 1, 1, x.size(3), device=x.device) > self.p
        return x * mask

# === 4. Attention L√©g√®re ===
class SimpleAttention(nn.Module):
    def __init__(self, in_channels):
        super().__init__()
        self.attn = nn.Sequential(
            nn.Conv2d(in_channels, 1, kernel_size=1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return x * self.attn(x)

# === DeepConvNet am√©lior√© ===
class DeepConvNet(nn.Module):
    def __init__(self, in_chans, n_classes, input_time_length):
        super().__init__()
        self.conv1 = nn.Sequential(
            nn.Conv2d(1, 25, kernel_size=(1, 5)),
            nn.Conv2d(25, 25, kernel_size=(in_chans, 1)),
            nn.BatchNorm2d(25),
            nn.ELU(),
            nn.MaxPool2d(kernel_size=(1, 2)),
            nn.Dropout(0.5)
        )
        self.conv2 = nn.Sequential(
            nn.Conv2d(25, 50, kernel_size=(1, 5)),
            nn.BatchNorm2d(50),
            nn.ELU(),
            nn.MaxPool2d(kernel_size=(1, 2)),
            nn.Dropout(0.5)
        )
        self.conv3 = nn.Sequential(
            nn.Conv2d(50, 100, kernel_size=(1, 5)),
            nn.BatchNorm2d(100),
            nn.ELU(),
            nn.MaxPool2d(kernel_size=(1, 2)),
            nn.Dropout(0.5)
        )
        self.conv4 = nn.Sequential(
            nn.Conv2d(100, 200, kernel_size=(1, 5)),
            nn.BatchNorm2d(200),
            nn.ELU(),
            TemporalDropout(p=0.3),
            SimpleAttention(200),
            nn.MaxPool2d(kernel_size=(1, 2)),
            nn.Dropout(0.5)
        )
        with torch.no_grad():
            dummy_input = torch.zeros(1, 1, in_chans, input_time_length)
            out = self._forward_features(dummy_input)
            self.flat_dim = out.view(1, -1).shape[1]
        self.classifier = nn.Linear(self.flat_dim, n_classes)

    def _forward_features(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.conv3(x)
        x = self.conv4(x)
        return x

    def forward(self, x):
        if x.ndim == 3:
            x = x.unsqueeze(1)
        x = self._forward_features(x)
        x = x.view(x.size(0), -1)
        return self.classifier(x)

# === Cross-validation stratifi√©e ===
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)

for fold, (train_idx, test_idx) in enumerate(kf.split(X, y)):
    print(f"\n===== Fold {fold+1} =====")
    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]

    train_loader = DataLoader(TensorDataset(X_train, y_train), batch_size=64, shuffle=True)

    model = DeepConvNet(in_chans=X.shape[1], n_classes=n_classes, input_time_length=X.shape[2]).to(DEVICE)
    criterion = FocalLoss(gamma=2.0, weight=weights_tensor)
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

    for epoch in range(1, 11):
        model.train()
        total_loss = 0
        correct = 0
        for xb, yb in train_loader:
            xb, yb = xb.to(DEVICE), yb.to(DEVICE)
            xb, yb_a, yb_b, lam = mixup_data(xb, yb)
            optimizer.zero_grad()
            out = model(xb)
            loss = mixup_criterion(criterion, out, yb_a, yb_b, lam)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
            pred = out.argmax(dim=1)
            correct += (pred == yb).sum().item()
        acc = correct / len(y_train)
        print(f"üß† Fold {fold+1} ‚Äî Epoch {epoch:02d} ‚Äî Loss: {total_loss:.4f} ‚Äî Acc: {acc*100:.2f}%")

    model.eval()
    with torch.no_grad():
        y_pred = model(X_test.to(DEVICE)).argmax(dim=1).cpu()
        print(classification_report(y_test, y_pred, digits=3, zero_division=0))
        cm = confusion_matrix(y_test, y_pred)
        sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
        plt.title(f"Fold {fold+1} ‚Äî Confusion Matrix")
        plt.xlabel("Predicted")
        plt.ylabel("True")
        plt.show()
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
from torch.utils.data import TensorDataset, DataLoader
from sklearn.model_selection import StratifiedKFold
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# === Chargement des donn√©es ===
X = np.load("X_windows.npy")
y = np.load("y_windows.npy")

# Correction des labels (0-index)
unique_labels = np.unique(y)
label_map = {old: new for new, old in enumerate(unique_labels)}
y = np.array([label_map[label] for label in y])
n_classes = len(unique_labels)

# Conversion en tensors
X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.long)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
SEED = 42
np.random.seed(SEED)
torch.manual_seed(SEED)

# === Class weights pour Focal Loss ===
class_weights = compute_class_weight(class_weight='balanced', classes=np.unique(y), y=y.numpy())
weights_tensor = torch.tensor(class_weights, dtype=torch.float32).to(DEVICE)

# === Modules avanc√©s ===
class FocalLoss(nn.Module):
    def __init__(self, gamma=2.0, weight=None):
        super().__init__()
        self.gamma = gamma
        self.weight = weight

    def forward(self, input, target):
        logpt = F.log_softmax(input, dim=1)
        pt = torch.exp(logpt)
        logpt = (1 - pt) ** self.gamma * logpt
        loss = F.nll_loss(logpt, target, weight=self.weight)
        return loss

class TemporalDropout(nn.Module):
    def __init__(self, p=0.3):
        super().__init__()
        self.p = p

    def forward(self, x):
        if not self.training or self.p == 0:
            return x
        mask = torch.rand(x.size(0), 1, 1, x.size(3), device=x.device) > self.p
        return x * mask

class SimpleAttention(nn.Module):
    def __init__(self, in_channels):
        super().__init__()
        self.attn = nn.Sequential(
            nn.Conv2d(in_channels, 1, kernel_size=1),
            nn.Sigmoid()
        )

    def forward(self, x):
        w = self.attn(x)
        return x * w

# === Mixup classique ===
def mixup_data(x, y, alpha=0.4):
    lam = np.random.beta(alpha, alpha) if alpha > 0 else 1
    batch_size = x.size(0)
    index = torch.randperm(batch_size).to(x.device)
    mixed_x = lam * x + (1 - lam) * x[index, :]
    y_a, y_b = y, y[index]
    return mixed_x, y_a, y_b, lam

def mixup_criterion(criterion, pred, y_a, y_b, lam):
    return lam * criterion(pred, y_a) + (1 - lam) * criterion(pred, y_b)

# === DeepConvNet with Attention & TemporalDropout ===
class DeepConvNet(nn.Module):
    def __init__(self, in_chans, n_classes, input_time_length):
        super().__init__()
        self.conv1 = nn.Sequential(
            nn.Conv2d(1, 25, (1, 5)),
            nn.Conv2d(25, 25, (in_chans, 1)),
            nn.BatchNorm2d(25), nn.ELU(),
            nn.MaxPool2d((1, 2)), nn.Dropout(0.5)
        )
        self.conv2 = nn.Sequential(nn.Conv2d(25, 50, (1, 5)), nn.BatchNorm2d(50), nn.ELU(), nn.MaxPool2d((1, 2)), nn.Dropout(0.5))
        self.conv3 = nn.Sequential(nn.Conv2d(50, 100, (1, 5)), nn.BatchNorm2d(100), nn.ELU(), nn.MaxPool2d((1, 2)), nn.Dropout(0.5))
        self.conv4 = nn.Sequential(
            nn.Conv2d(100, 200, (1, 5)), nn.BatchNorm2d(200), nn.ELU(),
            TemporalDropout(p=0.3), SimpleAttention(200),
            nn.MaxPool2d((1, 2)), nn.Dropout(0.5))

        with torch.no_grad():
            dummy = torch.zeros(1, 1, in_chans, input_time_length)
            out = self._forward_features(dummy)
            self.flat_dim = out.view(1, -1).shape[1]

        self.classifier = nn.Linear(self.flat_dim, n_classes)

    def _forward_features(self, x):
        x = self.conv1(x); x = self.conv2(x); x = self.conv3(x); x = self.conv4(x)
        return x

    def forward(self, x):
        if x.ndim == 3: x = x.unsqueeze(1)
        x = self._forward_features(x)
        return self.classifier(x.view(x.size(0), -1))

# === Entra√Ænement avec Mixup + FocalLoss ===
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)
for fold, (train_idx, test_idx) in enumerate(kf.split(X, y)):
    print(f"\n===== Fold {fold+1} =====")
    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]
    train_loader = DataLoader(TensorDataset(X_train, y_train), batch_size=64, shuffle=True)

    model = DeepConvNet(in_chans=X.shape[1], n_classes=n_classes, input_time_length=X.shape[2]).to(DEVICE)
    criterion = FocalLoss(weight=weights_tensor)
    optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)

    for epoch in range(1, 31):
        model.train()
        total_loss, correct = 0, 0
        for xb, yb in train_loader:
            xb, yb = xb.to(DEVICE), yb.to(DEVICE)
            xb, yb_a, yb_b, lam = mixup_data(xb, yb, alpha=0.4)
            out = model(xb)
            loss = mixup_criterion(criterion, out, yb_a, yb_b, lam)
            optimizer.zero_grad(); loss.backward(); optimizer.step()
            total_loss += loss.item()
            correct += (lam * (out.argmax(1) == yb_a).sum().item() + (1 - lam) * (out.argmax(1) == yb_b).sum().item())

        acc = correct / len(train_loader.dataset)
        print(f"Fold {fold+1} ‚Äî Epoch {epoch:02d} ‚Äî Loss: {total_loss:.4f} ‚Äî Acc: {acc*100:.2f}%")

    model.eval()
    with torch.no_grad():
        y_pred = model(X_test.to(DEVICE)).argmax(dim=1).cpu()
        print(classification_report(y_test, y_pred, target_names=[str(k) for k in unique_labels], zero_division=0))
        cm = confusion_matrix(y_test, y_pred)
        sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
        plt.title(f"Fold {fold+1} ‚Äî Confusion Matrix")
        plt.xlabel("Predicted")
        plt.ylabel("True")
        plt.show()
```
